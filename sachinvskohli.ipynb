```python
# prompt: read csv

import pandas as pd

# Assuming the CSV file is named 'your_file.csv' and is in the current directory.
# Replace 'your_file.csv' with the actual file name if it's different.
try:
    df = pd.read_csv('ODI_data.csv')
    print(df)
except FileNotFoundError:
    print("Error: 'your_file.csv' not found. Please make sure the file exists in the current directory or provide the correct path.")
except pd.errors.EmptyDataError:
    print("Error: 'your_file.csv' is empty.")
except pd.errors.ParserError:
    print("Error: Unable to parse 'your_file.csv'. Please check the file format.")
```

```python
# runs per innings
# sr
# 100s
# 50s
# team contribution
df['Innings Runs Scored Num'].unique()
```

```python
df= df[df['Innings Runs Scored Num'] != '-']
df['Innings Runs Scored Num'].unique()
```

```python
df = df.dropna(subset=['Innings Runs Scored Num'])
df['Innings Runs Scored Num'].unique()
```

```python
df['Innings Date'] = pd.to_datetime(df['Innings Date'], dayfirst=True)
df['year'] = df['Innings Date'].dt.year
df


```

```python
sachin_df= df[(df.year>=1994) & (df.year<=2004)]
sachin_df.head()
```

```python
kohli_df= df[(df.year>=2009) & (df.year<=2019)]
kohli_df.head()
```

```python
# metrics
# RPI(runs per inning) = total runs/ total innings
# SR(strike rate) = total runs/ total balls faced
# 100s = count number of 100s
# 50s = count number of 50
# team contribution = player runs/team runs
```

```python
df['Innings Runs Scored Num'] = df['Innings Runs Scored Num'].astype(int)
```

```python
df['Innings Not Out Flag'] = df['Innings Not Out Flag'].astype(int)
```

```python
df['Innings balls Faced'] = df['Innings Balls Faced'].astype(int)
```

```python
sdf= sachin_df[sachin_df['Innings Player']=='SR Tendulkar']
sdf
```

```python
sum_runs = sdf['Innings Runs Scored Num'].sum()
sum_runs
```

```python
kdf= kohli_df[kohli_df['Innings Player']=='V Kohli']
kdf
```

```python
sum(kdf['Innings Runs Scored Num'])/len(kdf), sum(sdf['Innings Runs Scored Num'])/len(sdf)
```

```python
#SR
100*sum(kdf['Innings Runs Scored Num'])/sum(kdf['Innings balls Faced']), 100*sum(sdf['Innings Runs Scored Num'])/sum(sdf['Innings balls Faced'])
```
